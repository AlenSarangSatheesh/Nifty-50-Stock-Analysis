# Frequency of Opening Prices Exceeding Closing Prices


```python
def analyze_open_close(stock_df):
    open_close_df = pd.DataFrame()                     #creating dataframe
    open_close_df['Open'] = stock_df['Open']           #Creating 2 columns for opening closing prices
    open_close_df['Close'] = stock_df['Close']

    # Creating column having difference between Open and Close prices
    open_close_df['OpenCloseDifference'] = open_close_df['Open'] - open_close_df['Close']

    #total trading days
    total_trading_days = len(open_close_df)
    
    # count  Opening price higher than closing
    open_high_count = (open_close_df['OpenCloseDifference'] > 0).sum()
    
    # count  Opening price lesser than closing
    open_less_count = (open_close_df['OpenCloseDifference'] < 0).sum()
    
    #Opening price = Closing Price
    no_price_change = (open_close_df['OpenCloseDifference'] == 0).sum()
     
    #Percentages
    percent_open_high = (open_high_count / total_trading_days) * 100
    percent_open_less = (open_less_count / total_trading_days) * 100
    percent_no_change = (no_price_change / total_trading_days) * 100
    
    
    

    # Return the results
    return {
        'Open Price greater than Closing': open_high_count,
        'Closing price greater than Opening': open_less_count,
        'Opening price = Closing price': no_price_change,
        'Opening Price greater than closing %': percent_open_high,
        'Opening Price lesser than closing %': percent_open_less,
        'No price change %': percent_no_change
        
    }
```

```python
def print_analysis_results(filename):
    stock_data = pd.read_csv(filename)
    results = analyze_open_close(stock_data)
    
    stock_name = os.path.basename(filename).split(".")[0]  # Extracting the stock name from the filename
    
    print(f"{stock_name} Analysis Results:")
    print('-------------------------------------------------------------------------------------------------------------------')
    for key, value in results.items():
        print(f"{key}: {value}")
    print()

#storing file names as array
filenames = [
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ADANIPORTS.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ASIANPAINT.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/AXISBANK.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BAJAJ-AUTO.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BAJAJFINSV.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BAJFINANCE.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BHARTIARTL.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BPCL.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BRITANNIA.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/CIPLA.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/COALINDIA.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/DRREDDY.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/EICHERMOT.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/GAIL.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/GRASIM.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/HCLTECH.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/HDFC.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/HDFCBANK.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/HEROMOTOCO.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/HINDALCO.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/HINDUNILVR.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ICICIBANK.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/INDUSINDBK.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/INFY.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/IOC.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ITC.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/JSWSTEEL.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/KOTAKBANK.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/LT.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/MM.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/MARUTI.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/NESTLEIND.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/NTPC.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ONGC.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/POWERGRID.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/RELIANCE.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/SBIN.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/SHREECEM.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/SUNPHARMA.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/TATAMOTORS.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/TATASTEEL.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/TCS.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/TECHM.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/TITAN.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ULTRACEMCO.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/UPL.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/VEDL.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/WIPRO.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ZEEL.csv"
               ]
for filename in filenames:
    print_analysis_results(filename)
```

```python
def analyse_daily_return(stock_data):
    open_close_df = pd.DataFrame()                     #creating dataframe
    open_close_df['Open'] = stock_data['Open']           #Creating 2 columns for opening closing prices
    open_close_df['Close'] = stock_data['Close']
    
    #Daily return Dataframe column
    open_close_df['Daily-Return %'] = (open_close_df['Close'] - open_close_df['Open']) / open_close_df['Open']
    return open_close_df
```

```python

def load_stock_data(filenames):
    stock_data = {}
    for filename in filenames:
        stock = pd.read_csv(filename)
        stock_data[stock['Symbol'][0]] = stock['Close']
    return stock_data

def calculate_correlation_matrix(stock_data):
    df = pd.DataFrame(stock_data)
    correlation_matrix = df.corr()
    return correlation_matrix

filenames = [                                            #giving few data for readbility
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ADANIPORTS.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/ASIANPAINT.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/AXISBANK.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BAJAJ-AUTO.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BAJAJFINSV.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BAJFINANCE.csv",
              "C:/Users/pc/Desktop/Stock Market Prediction/DataBase/BHARTIARTL.csv",
            
               ]
stock_data = load_stock_data(filenames)
correlation_matrix = calculate_correlation_matrix(stock_data)

# Convert the correlation matrix to a numpy array of float values
correlation_array = correlation_matrix.values.astype(float)

plt.figure(figsize=(10, 8))
plt.imshow(correlation_array, cmap='viridis', interpolation='nearest', aspect='auto', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title("Correlation Matrix of Closing Prices")
plt.show()

#https://muthu.co/understanding-correlations-and-correlation-matrix/
#Study more , code not understandable . Copied from internet
```
